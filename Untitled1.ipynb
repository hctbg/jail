```python
from bs4 import BeautifulSoup
from datetime import timedelta
import requests
import requests_cache
from time import sleep
from peewee import DoesNotExist
from urllib.parse import parse_qs, urlparse
```

```python
r = requests.get(
    'https://report.boonecountymo.org/mrcjava/servlet/RMS01_MP.R00040s?run=2&R001=&R002=&ID=3641&hover_redir=&width=950',
    headers={'user-agent': "I'm good people!!!"}
)
print(r.request.headers)
print(r.status_code)
print(r.reason)
```

```python
with open ('homepage.html','w') as f:
    f.write(r.text)
```

```python
soup = BeautifulSoup(r.text)
```

```python
tables = soup.find('div', id = 'content')
```

```python
td_all = tables.find_all('td', class_='two td_left') #td means all detail infos and element 
```

```python
for td in td_all:
    print(td.text)
```

```python
#all IDs 

url = soup.select("div#content > div",recursive=False)

for tag in url:
    div_ID = tag.get('id')
    ids = div_ID.lstrip('mugshot')
    print(ids)
```

```python
#for all names


all_names = soup.find_all('div', class_="inmateName")

detainee_name = []

for name in all_names:
    print(name.text.strip())
    
detainee_name.append(name)
```

```python
print(detainee_name)
```

```python
#for all detainee personal info
all_detail_info = tables.find_all('table', class_="collapse centered_table shadow")
for td in all_detail_info:
    for td in td.find_all('td',class_='two td_left'):
        print(td.text.strip())
    print('---------------')
```
